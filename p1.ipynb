```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
employee = pd.read_excel("dane/TTex_2023_t06.xlsx", sheet_name="Employee")
department = pd.read_excel("dane/TTex_2023_t06.xlsx", sheet_name="Department")
position = pd.read_excel("dane/TTex_2023_t06.xlsx", sheet_name="Positon")
car = pd.read_excel("dane/TTex_2023_t06.xlsx", sheet_name="Car")
trip = pd.read_excel("dane/TTex_2023_t06.xlsx", sheet_name="Trip")
```

```python
wyjazdy = trip.groupby(["employee"]).size().reset_index(name="wyjazdy")
wyjazdy_pracownik = pd.merge(wyjazdy, employee, left_on= "employee", right_on="id")
print(wyjazdy_pracownik[['surname', 'name', 'wyjazdy']])
```

```python
paliwo_pracownik = trip[["employee", "fuel", "distance"]]
department_pracownik = employee[["id", "department"]]
paliwo_pracownik = pd.merge(paliwo_pracownik, department_pracownik, left_on="employee", right_on="id")
paliwo_pracownik = pd.merge(paliwo_pracownik, department, left_on="department", right_on="id")
paliwo_department = paliwo_pracownik[["department","name","description","fuel","distance"]].groupby(["department","name","description"]).sum().reset_index()
# print(paliwo_pracownik)
paliwo_department["l/100km"] = paliwo_department["fuel"] / paliwo_department["distance"] * 100
print(paliwo_department[["description","l/100km"]].sort_values("l/100km", ascending=False))
```

```python
# Sortowanie danych według wartości "l/100km"
paliwo_department_sorted = paliwo_department.sort_values("l/100km", ascending=False)

# Tworzenie wykresu słupkowego
plt.figure(figsize=(10, 6))
plt.bar(paliwo_department_sorted["description"], paliwo_department_sorted["l/100km"], color='skyblue')

# Dodanie tytułu i etykiet osi
plt.title("Zużycie paliwa (l/100km) według działu", fontsize=14)
plt.xlabel("Dział", fontsize=12)
plt.ylabel("Zużycie paliwa (l/100km)", fontsize=12)
plt.xticks(rotation=45, ha='right')  # Obrót etykiet na osi X dla lepszej czytelności

# Wyświetlenie wykresu
plt.tight_layout()
plt.show()
```

```python
def load(startWeek, endWeek):
    files = [f"dane/TTex_2023_t{m:02d}.xlsx" for m in range(startWeek, endWeek+1)]
    t_list = [pd.read_excel(f, sheet_name="Trip") for f in files]
    trip = pd.concat(t_list, ignore_index=True)
    return trip

trip = load(3,6)
```
